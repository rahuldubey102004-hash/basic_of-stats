# Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading DataSet into notebook
Data=pd.read_csv("sales_data_with_discounts.csv")
```

```python
Data.head()
```

```python
Data.info()
```

```python
#Identifying numerical columns through select_dtypes and forming them into list for access
numerical_cols=list(Data.select_dtypes(include='number').columns)
numerical_cols
```

## 1.Descriptive Analytics for Numerical columns

```python
# find mean,median,mode and standard deviation for Numerical Columns
```

```python
for col in numerical_cols:
    print("\n")
    print("-"*40)
    print(f"Column :-{col}")
    print(f"Mean  :{Data[col].mean():.3f}")
    print(f"Median :{Data[col].median():.3f}")
    print(f"Mode  {Data[col].mode()[0]:.3f}")
    print(f"Standarad Deviation {Data[col].std():.3f}")
    print("Interpretation:-")
    print(f"The Average of the {col} column is {Data[col].mean():.3f}")
    print(f"The Median of the {col} column is {Data[col].median():.3f}")
    print(f"The Most repeating element  of the {col} column is {Data[col].mode()[0]:.3f}")
    print(f"The Spread of the Data i.e Standard deviation of the {col} column is {Data[col].std():.3f}")
    print("-"*40)
```

## 2.Data Visualization

```python
# plotting histogram for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(6,4),dpi=100,facecolor='lightgrey')
    sns.histplot(Data[col],kde=True)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

#Analysis of the histograms
    print(f"Analysis of {col}:")
    print(f"Skewness of {col} is {Data[col].skew()}")
    print(f"kurtosis of {col} is {Data[col].kurt()}")

#printing Analysis
    if Data[col].skew()==0:
        print(f"{col} is Perfectly Skewed ")
    elif Data[col].skew()>0:
        print(f"{col} is right skewed")
    else:
        print(f"{col} is left skewed")
```

```python
# Box Plot Plotting
for col in numerical_cols:
    plt.figure(figsize=(4,4),facecolor='lightgrey')
    box=plt.boxplot(Data[col])
    plt.ylabel(col)
```

```python
categorical_cols=list(Data.select_dtypes(include='object').columns)
categorical_cols
```

```python
for col in categorical_cols:
    plt.figure(figsize=(10,6))
    sns.countplot(data=Data,x=col,order=Data[col].value_counts().index)
    plt.title(f'Count of each categories of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
